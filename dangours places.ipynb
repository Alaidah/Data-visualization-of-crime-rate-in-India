```python
crime1=pd.read_csv('17_Case_reported_and_value_of_property_taken_away_by_place_of_occurrence_2001_2012.csv')
crime2=pd.read_csv('17_Case_reported_and_value_of_property_taken_away_by_place_of_occurrence_2013.csv')

crime_place = pd.concat([crime1,crime2],  ignore_index=False, axis=0)
```

```python
del crime1
del crime2
crime_place['Theft (Section 379-382 IPC) - Number of cases registered']=crime_place['Theft (Section 379-382 IPC) - Number of cases registered']+crime_place['Dacoity (Section 395-398 IPC) - Number of cases registered']+crime_place['Robbery(Section 392-394, 397, 398 IPC) - Number of cases registered']+crime_place['Burglary(Section 449-452, 454, 455, 457-460 IPC) - Number of cases registered']
crime_place['Theft (Section 379-382 IPC) - Value Of Property Stolen (in rupees)']=crime_place['Theft (Section 379-382 IPC) - Value Of Property Stolen (in rupees)']+crime_place['Dacoity (Section 395-398 IPC) - Value Of Property Stolen (in rupees)']+crime_place['Burglary(Section 449-452, 454, 455, 457-460 IPC) - Value Of Property Stolen (in rupees)']+crime_place['Robbery(Section 392-394, 397, 398 IPC) - Value Of Property Stolen (in rupees)']
```

```python
del crime_place['Dacoity (Section 395-398 IPC) - Number of cases registered']
del crime_place['Dacoity (Section 395-398 IPC) - Value Of Property Stolen (in rupees)']
del crime_place['Robbery(Section 392-394, 397, 398 IPC) - Number of cases registered']
del crime_place['Robbery(Section 392-394, 397, 398 IPC) - Value Of Property Stolen (in rupees)']
del crime_place['Burglary(Section 449-452, 454, 455, 457-460 IPC) - Number of cases registered']
del crime_place['Burglary(Section 449-452, 454, 455, 457-460 IPC) - Value Of Property Stolen (in rupees)']
```

```python
crime_place.head()
```

```python
crime_place= crime_place[(crime_place['Place Of Occurrence']== 'RESIDENTIAL PREMISES') | (crime_place['Place Of Occurrence']== 'HIGH-WAY')|(crime_place['Place Of Occurrence']=='RIVER & SEA')|(crime_place['Place Of Occurrence']=='RUNNING TRAINS')|(crime_place['Place Of Occurrence']=='COMMERCIAL ESTABLISHMENTS' )| (crime_place['Place Of Occurrence']=='BANKS')]
```

```python
crime_place
```

```python
crime_place_g= crime_place.groupby(['STATE/UT','Place Of Occurrence']).sum().reset_index()
```

```python
crime_place_g.rename(columns={'Theft (Section 379-382 IPC) - Number of cases registered':'Theft cases registered'}, inplace=True)
```

```python
crime_place_g.rename(columns={'Theft (Section 379-382 IPC) - Value Of Property Stolen (in rupees)':'Value Of Property Stolen (in rupees)'}, inplace=True)
```

```python
crime_place_g['Place Of Occurrence'] =crime_place_g['Place Of Occurrence'].str.lower()
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

from plotly.tools import FigureFactory as FF
import scipy

import dash
import dash_core_components as dcc
import dash_html_components as html

```

```python
#crime_place_g['Theft cases registered']=((crime_place_g['Theft cases registered']-crime_place_g['Theft cases registered'].min())/(crime_place_g['Theft cases registered'].max()-crime_place_g['Theft cases registered'].min()))

crime_place_g['Text'] = "<br>Place Of Occurrence: "+crime_place_g['Place Of Occurrence'] + "<br>STATE: "+crime_place_g['STATE/UT']+"<br>cases registered: "+" ("+crime_place_g['Theft cases registered'].astype(str)+")"+"<br>Value Of Property Stolen (in rupees):" +" ("+crime_place_g['Value Of Property Stolen (in rupees)'].astype(str)+")" 
trace = go.Heatmap(z=crime_place_g['Theft cases registered'],
                   x=crime_place_g['Place Of Occurrence'],
                   y=crime_place_g['STATE/UT'],
                   colorscale='Reds',
                   text =crime_place_g['Text'],
                   hoverinfo='text')



layout = go.Layout(
   
    title='Most Dangerous Places Regarding Theft Crime In Indian states from 2001 to 2013 ',
   titlefont=dict(
            family='Courier New, monospace',
            size=19,
            color='rgba(152, 0, 0, .8)'
        
        ),
    
    yaxis=dict(
    
        tickmode='linear',
        categoryorder='array',
        titlefont=dict(
            family='Courier New, monospace',
            size=13,
            color='#7f7f7f'
        )
        
        
    ),
    
    
    xaxis=dict(
        title=' Place Of Occurance',
        titlefont=dict(
            family='Courier New, monospace',
            size=15,
            color='rgba(152, 0, 0, .8)'
        
        )
        
        
       
    ),
    
    autosize=False,
    width=900,
    height=800,
    margin=go.layout.Margin(
        l=150,
        r=80,
        b=150,
        t=70,
        pad=5
    )

    
)

data=[trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='marker-h-bar')

#py.iplot(data, filename='labelled-heatmap')
```
